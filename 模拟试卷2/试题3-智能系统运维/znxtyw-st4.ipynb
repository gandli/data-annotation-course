1.库文件导入

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

2.导入数据集

```python
(train_image,train_label),(test_image,test_label) = tf.keras.datasets.mnist.load_data()
```

```python
train_image.shape  , train_label.shape  ,test_image.shape , test_label.shape
```

```python
plt.figure(figsize=(2,2))
plt.imshow(train_image[1])
print(f"当前标签数字为{train_label[1]}，图形显示如下：")
```

```python
#热力图的方式
plt.figure(figsize=(12,12))
plt.imshow(train_image[0] , cmap='plasma')  #颜色的对比
for i in range(28):
    for j in range(28):
        text = plt.text(j, i, train_image[0][i, j] , c = 'g',ha = 'center')  
```

```python
#归一化
train_image = train_image/255.0
test_image  = test_image/255.0
```

3.模型搭建

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape = (28,28)))  #展开输入层
model.add(tf.keras.layers.Dense(64 , activation= tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(32 , activation= tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(10 , activation= tf.keras.activations.softmax))
model.summary()
```

4.训练配置

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  
              loss='sparse_categorical_crossentropy',  #稀疏交叉熵
              metrics=['acc']  #准确率
)
```

5.训练

```python
history = model.fit(train_image,train_label, epochs=10, validation_split = 0.2)
```

```python
plt.figure(figsize=(4, 4))
plt.plot(history.epoch, history.history.get('loss'), label='loss')
plt.plot(history.epoch, history.history.get('val_loss'), label='val_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('loss curve of intelligent system')
plt.legend()
```

```python
plt.figure(figsize=(4, 4))
plt.plot(history.epoch, history.history.get('acc'), label='acc')
plt.plot(history.epoch, history.history.get('val_acc'), label='val_acc')
plt.xlabel('epoch')
plt.ylabel('acc')
plt.title('acc curve of intelligent system')
plt.legend()
```

6.模型评估与预测

```python
loss, acc = model.evaluate(test_image, test_label)
print("智能系统运行")
print(f"模型准确率acc为{acc:.2%}")
print(f"模型损失loss为{loss:.2f}")
```

```python
def pre_fun(img):
    plt.figure(figsize=(2,2))
    plt.imshow(img, cmap='plasma')
    image = tf.expand_dims(img , 0) #扩增维度
    pre = model.predict(image)
    print('当前图片为：',np.argmax(pre))
```
